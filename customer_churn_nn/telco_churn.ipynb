```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
telco_cust = pd.read_csv('telco_dataset.csv')
telco_cust.head()
```

```python
telco_cust.columns.values
```

```python
# explore if there are missing values
telco_cust.dtypes
```

```python
# converting total charges to numerical data types
telco_cust.TotalCharges = pd.to_numeric(
    telco_cust.TotalCharges, errors="coerce")
telco_cust.isnull()
```

```python
#  remove missing values
telco_cust.dropna(inplace=True)
# remove customerID
telco_cust.drop('customerID', axis=1, inplace=True)
telco_cust
```

```python
# change male to female
telco_cust['gender'].replace({'Male': 1, 'Female': 0}, inplace=True)

telco_cust.replace("No internet service", "No", inplace=True)
telco_cust.replace("No phone service", "No", inplace=True)

# change (yes, no) to (1,0)
yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines',
                  'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']
for col in yes_no_columns:
    telco_cust[col].replace({'Yes': 1, 'No': 0}, inplace=True)

telco_cust
```

```python
# one hot encoding
data = pd.get_dummies(data=telco_cust, columns=[
                      'MultipleLines', 'InternetService', 'Contract', 'PaymentMethod'])

data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# scale all variables to range of 0 to 1
features = data.copy().columns.values
scaler.fit(data)
data = pd.DataFrame(scaler.transform(data))
data.columns = features
data.sample(5)
```

```python
X = data.drop('Churn', axis=1)
y = data['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=25)
```

```python
num_inputs = len(X_train.columns)
num_inputs
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(num_inputs, input_shape=(
        num_inputs,), activation="relu"),
    keras.layers.Dense(num_inputs,  activation="relu"),
    keras.layers.Dense(num_inputs, activation="relu"),
    keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```

```python
y_predict = model.predict(x_test)
y_predict > 0.5
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict > 0.5))
```
